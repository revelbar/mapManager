```python
import numpy as np
import matplotlib.pyplot as plt
import mapManager as mm
```

Lets open an image of a random place in US by GeoEye-1.

```python
data = mm.Image('geoeye-dc.jpg')
```

We should check the stats of the image before we do any sort of calculations

```python
data.info()
```

Everything here seems to be in order; the datatype matches with the range of values. If I try to display it:

```python
data.display()
```

Since everything is in order, I can safely make a map out of this image! 

The package has two types of bands: Grayscale and Multispectral. Grayscale maps handle the images that only have 1 band, while Multispectral maps handle those with more than 1 band.

Our image has three bands right now, so the following function will spit out a Multispectral map:

```python
sample_map = data.make_map('sample', 0.41)
```

```python
plt.figure(figsize=(10,10)) # for making the image bigger
sample_map.display()
```

If I only want to see the 1st band, I can do the following:

```python
plt.figure(figsize=(10,10)) # for making the image bigger
sample_map.display(1)
```

Here is a more detailed descriptive statistics on the map:

```python
print(sample_map.statistics())
```

Drawing the distribution of values per band, we get this:

```python
x=sample_map.histogram()
```

We can also increase the contrast of the image for a better look at the different features of the map.

```python
x = sample_map.inc_contrast()
```
